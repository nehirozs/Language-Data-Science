# Part 1: N-Gram Language Model + Decoding Strategies (45 points)

## Background

An **n-gram language model** assigns probability to a sequence of words by decomposing
it into conditional probabilities:

$$P(w_1, w_2, \ldots, w_T) = \prod_{t=1}^{T} P(w_t \mid w_{t-n+1}, \ldots, w_{t-1})$$

Each word depends only on the preceding **n − 1** words (the *context*).

### Padding with `<BOS>` and `<EOS>`

To handle sentence boundaries, we wrap every training sentence with special tokens:

```
<BOS>  the  cat  sat  on  the  mat  <EOS>
```

* `<BOS>` (beginning-of-sentence) fills in the left context for the first real word.
* `<EOS>` (end-of-sentence) is predicted as a regular token; generating it terminates decoding.

For a bigram model (n = 2) we prepend **one** `<BOS>` token; for a trigram model **two**, and so on.

### Laplace (Add-1) Smoothing

Raw MLE probabilities assign zero to any unseen n-gram. We use add-1 smoothing:

$$P_{\text{Laplace}}(w \mid \text{ctx}) = \frac{\text{count}(\text{ctx}, w) + 1}{\text{count}(\text{ctx}) + |V|}$$

where $|V|$ is the vocabulary size (tokens that can be predicted).

### Decoding Strategies

Once the model is trained, we can **generate** text from a prompt using each of the following decoding strategies:

| Strategy | Picks next token by… |
|---|---|
| **Greedy** | argmax probability at every step |
| **Beam search** | maintains `beam_width` hypotheses, maximises total log-prob |
| **Top-k sampling** | samples from the top-k most probable tokens (stochastic) |

All three strategies stop when `<EOS>` is generated or `max_length` tokens are reached.


````python
# --- Setup for Part 1 (provided, not graded — run this cell first) ---
import math, re
import numpy as np
import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize, word_tokenize
from collections import defaultdict

np.random.seed(345)

BOS = "<BOS>"
EOS = "<EOS>"

# Loading Moby Dick corpus
nltk.download("gutenberg", quiet=True)
nltk.download("punkt", quiet=True)
nltk.download("punkt_tab", quiet=True)

raw_text = gutenberg.raw("melville-moby_dick.txt")
sentences = sent_tokenize(raw_text)

CORPUS = []
for sent in sentences:
    toks = [t.lower() for t in word_tokenize(sent)]
    if toks:
        CORPUS.append(toks)

def tokenize(text):
    """Lowercase and split on whitespace/punctuation."""
    return re.findall(r"[a-z']+", text.lower())

def pad_sentence(tokens, n):
    """Wrap a token list with (n-1) <BOS> tokens on the left and one <EOS> on the right."""
    return [BOS] * (n - 1) + tokens + [EOS]

print(f"Corpus loaded (Moby Dick): {len(CORPUS)} sentences")
print(f"Example padded sentence (n=2): {pad_sentence(CORPUS[5], n=2)[:20]} ...")
````

````python
# --- Part 1 Examples (provided, not graded) ---
# Demonstrate counts/probabilities for multiple n values.

from collections import Counter

for n_ex in [2, 3, 4]:
    ex_counts = Counter()
    ex_ctx_counts = Counter()
    ex_vocab = set()

    for sent in CORPUS:
        ex_vocab.update(sent)
        padded = pad_sentence(sent, n_ex)
        for i in range(n_ex - 1, len(padded)):
            ctx  = tuple(padded[i - (n_ex - 1) : i])
            word = padded[i]
            ex_counts[ctx + (word,)] += 1
            ex_ctx_counts[ctx]       += 1

    ex_vocab.add(EOS)
    V = len(ex_vocab)

    print(f"n={n_ex}: vocab_size={V}, unique {n_ex}-grams={len(ex_counts)}")
    sample = list(ex_counts.items())[:3]
    for ng, cnt in sample:
        ctx = ng[:-1]
        w = ng[-1]
        ctx_cnt = ex_ctx_counts.get(ctx, 0)
        p = (cnt + 1) / (ctx_cnt + V)
        print(f"  count{ng}={cnt}  |  P({w} | {ctx})={p:.6f}")
    print("-" * 50)
````

### Q1.1 — `train_ngram` (9 pts)

Implement `train_ngram(corpus, n) -> dict`.

**What it must do:**
- For each sentence in `corpus`, pad it using the `pad_sentence(sent, n)` defined above to add
  `(n-1)` `<BOS>` tokens on the left and one `<EOS>` on the right.
- Count every n-gram and every (n-1)-gram context found in the padded sentences.
- Build the vocabulary: all unique tokens that appear in the raw (unpadded) sentences **plus** `<EOS>`.
  (Hint: `<BOS>` is **never** in the vocab because it is never predicted.)

**Return** a dictionary with exactly these keys:

| Key | Type | Description |
|---|---|---|
| `"n"` | `int` | The n-gram order |
| `"counts"` | `dict` | Maps `(ctx_token, ..., word)` → `int` count |
| `"context_counts"` | `dict` | Maps `(ctx_token, ...)` → `int` (total continuations seen for that context) |
| `"vocab"` | `set` | All predictable tokens (no `<BOS>`) |
| `"vocab_size"` | `int` | `len(vocab)` |

**Hint:** At each position, look back (n-1) words to form the context, pair it with the current word, and update the counts.


````python
def train_ngram(corpus, n):
    """
    Train an n-gram language model with Laplace smoothing.

    Args:
        corpus: List of sentences; each sentence is a list of string tokens.
        n:      N-gram order (e.g. 2 = bigram, 3 = trigram).

    Returns:
        dict with keys: 'n', 'counts', 'context_counts', 'vocab', 'vocab_size'.
    """
    counts = defaultdict(int)
    context_counts = defaultdict(int)
    vocab = set()
    for sent in corpus:
        vocab.update(sent)
        padded = pad_sentence(sent, n)
        for i in range(n - 1, len(padded)):
            ctx = tuple(padded[i - (n - 1) : i])
            word = padded[i]
            counts[ctx + (word,)] += 1
            context_counts[ctx] += 1
    vocab.add(EOS)
    return {
        "n": n,
        "counts": dict(counts),
        "context_counts": dict(context_counts),
        "vocab": vocab,
        "vocab_size": len(vocab),
    }
````

````python
# --- Check your work: Q1.1 ---
model_bi  = train_ngram(CORPUS, n=2)
model_tri = train_ngram(CORPUS, n=3)

print(f"Bigram model:  vocab_size={model_bi['vocab_size']},  "
      f"unique bigrams={len(model_bi['counts'])}")
print(f"Trigram model: vocab_size={model_tri['vocab_size']}, "
      f"unique trigrams={len(model_tri['counts'])}")

# BOS should NOT be in vocab
assert BOS not in model_bi["vocab"], "BOS should not be in vocab"

# EOS should be in vocab
assert EOS in model_bi["vocab"], "EOS should be in vocab"

# A known bigram count
bc = model_bi["counts"].get((BOS, "the"), 0)
print(f"count(<BOS>, 'the') = {bc}  (should be > 0)")
assert bc > 0

# Context count for <BOS> should equal number of training sentences
bos_ctx = model_bi["context_counts"].get((BOS,), 0)
print(f"context_count(<BOS>) = {bos_ctx}  (should equal {len(CORPUS)})")
assert bos_ctx == len(CORPUS), f"Got {bos_ctx}"

print("\nAll checks passed!")
````

### Q1.2 — `score_sequence` (9 pts)

Implement `score_sequence(model, tokens) -> float`.

Compute the **log-probability** of a token sequence under the trained n-gram model.

**Requirements:**
- `tokens` is a plain list of strings (no padding).
- Internally, prepend `(n-1)` `<BOS>` tokens to create the full context for position 0.
- For each token at position `i` (0-indexed in `tokens`):
  - Context: `ctx = tuple(padded[i : i + n - 1])`  where `padded = [BOS]*(n-1) + tokens`
  - Apply **Laplace smoothing**:
    `prob = (counts.get(ctx + (token,), 0) + 1) / (context_counts.get(ctx, 0) + vocab_size)`
  - Add `math.log(prob)` to the running total.
- Return the total log-probability as a `float`.

**Note:** Do **not** include the `<EOS>` token in `tokens`; `score_sequence` scores the
observable words only.


````python
def score_sequence(model, tokens):
    """
    Compute the log-probability of a token sequence under the n-gram model.

    Args:
        model:  Trained model dict returned by train_ngram().
        tokens: List of string tokens to score (no BOS/EOS padding needed).

    Returns:
        Total log-probability (float, always <= 0).
    """
    n = model["n"]
    padded = [BOS] * (n - 1) + list(tokens)
    total = 0.0
    V = model["vocab_size"]
    counts = model["counts"]
    context_counts = model["context_counts"]
    for i, tok in enumerate(tokens):
        ctx = tuple(padded[i : i + n - 1])
        prob = (counts.get(ctx + (tok,), 0) + 1) / (context_counts.get(ctx, 0) + V)
        total += math.log(prob)
    return total
````

````python
# --- Check your work: Q1.2 ---
model_bi = train_ngram(CORPUS, n=2)

# In-corpus tokens should score better than random ones
lp_known   = score_sequence(model_bi, ["the", "cat", "sat"])
lp_unknown = score_sequence(model_bi, ["xyzzy", "foo", "bar"])

print(f"log P(['the','cat','sat'])   = {lp_known:.4f}")
print(f"log P(['xyzzy','foo','bar']) = {lp_unknown:.4f}")
assert lp_known > lp_unknown, "In-corpus sequence should have higher log-prob"

# Log-prob must be <= 0
assert lp_known <= 0 and lp_unknown <= 0, "Log-probs must be non-positive"

# Single-token check: manual Laplace bigram probability
ctx = (BOS,)
word = "the"
manual = (model_bi["counts"].get(ctx + (word,), 0) + 1) / (model_bi["context_counts"].get(ctx, 0) + model_bi["vocab_size"])
assert abs(score_sequence(model_bi, ["the"]) - math.log(manual)) < 1e-9, "Manual Laplace mismatch"

print("\nAll checks passed!")
````

### Q1.3 — `greedy_decode` (9 pts)

**What greedy decoding does:** at each generation step, it commits to the single
highest-probability next token and never revisits that decision later.

**How to think about the implementation:**
- Tokenize the prompt and build the current context from the last `n-1` tokens.
- Score every token in the vocabulary under that context.
- Pick the token with the largest probability.
- Append it to the output, update the context window, and repeat.
- Stop if `<EOS>` is selected or when `max_length` is reached.

**Toy example:**
- Suppose the current context is `("the",)` and the model gives:
  `P(cat | the)=0.50`, `P(dog | the)=0.30`, `P(mat | the)=0.20`
- Greedy decoding picks `cat` immediately because it has the largest probability.
- If the next context becomes `("cat",)` and `<EOS>` has the highest probability,
  generation stops there.

**References:**
- Hugging Face text generation overview: https://huggingface.co/blog/how-to-generate#greedy-search
- Hugging Face generation strategies docs: https://huggingface.co/docs/transformers/main/en/generation_strategies#greedy-search


````python
def greedy_decode(model, prompt, max_length = 20):
    """
    Generate text from a prompt using greedy decoding.

    Args:
        model:      Trained model dict returned by train_ngram().
        prompt:     Input text string (will be tokenized).
        max_length: Maximum number of new tokens to generate.

    Returns:
        Full string: original prompt tokens + generated tokens, space-joined.
    """
    n = model["n"]
    output = list(tokenize(prompt))
    vocab = sorted(model["vocab"])
    V = model["vocab_size"]
    counts = model["counts"]
    context_counts = model["context_counts"]

    def gen_context(seq):
        need = n - 1
        if len(seq) >= need:
            return tuple(seq[-need:])
        return tuple([BOS] * (need - len(seq)) + seq)

    for _ in range(max_length):
        ctx = gen_context(output)
        best_tok = None
        best_p = -1.0
        for w in vocab:
            p = (counts.get(ctx + (w,), 0) + 1) / (context_counts.get(ctx, 0) + V)
            if p > best_p:
                best_p = p
                best_tok = w
        if best_tok == EOS:
            break
        output.append(best_tok)
    return " ".join(output)
````

````python
# --- Check your work: Q1.3 ---
model_bi  = train_ngram(CORPUS, n=2)
model_tri = train_ngram(CORPUS, n=3)

for n_label, model in [("bigram", model_bi), ("trigram", model_tri)]:
    out = greedy_decode(model, "the cat", max_length=10)
    print(f"greedy_decode({n_label}, 'the cat'): {out!r}")
    assert out.startswith("the cat"), "Output must begin with the prompt tokens"

# k=1 in top-k should match greedy (tested again in Q1.5)
print("\nAll checks passed!")
````

### Q1.4 — `beam_decode` (9 pts)

Implement `beam_decode(model, prompt, beam_width=3, max_length=20) -> str`.

**What beam search does:** instead of keeping only one partial continuation, it keeps
the best few candidates at each step, measured by cumulative log-probability.

**How to think about the implementation:**
- Start with one empty continuation and the prompt context.
- For every active beam, try every possible next token.
- Add `log(prob)` to the beam score so longer sequences are scored by the sum of
  their token log-probabilities.
- Keep only the top `beam_width` unfinished beams for the next round.
- If a beam emits `<EOS>`, move it to a list of completed candidates.
- At the end, return the best candidate among the completed or remaining beams.

**Toy example:**
- Suppose after the prompt `"the"` the two best next tokens are `cat` and `dog`.
- With `beam_width=2`, you keep both partial sequences: `"the cat"` and `"the dog"`.
- On the next step, `"the dog barked"` might outrank `"the cat sat"` overall even if
  `cat` was the best first choice locally.
- This is the main difference from greedy decoding.

**Constraint:** `beam_width=1` must produce the **same output** as `greedy_decode`.

**References:**
- Hugging Face text generation overview: https://huggingface.co/blog/how-to-generate#beam-search
- Beam search discussion in Transformers docs: https://huggingface.co/docs/transformers/main/en/generation_strategies#beam-search


````python
def beam_decode(model, prompt, beam_width = 3, max_length = 20):
    """
    Generate text from a prompt using beam search.

    Args:
        model:      Trained model dict returned by train_ngram().
        prompt:     Input text string (will be tokenized).
        beam_width: Number of beams to keep at each step.
        max_length: Maximum number of new tokens to generate.

    Returns:
        Full string: original prompt tokens + generated tokens, space-joined.
    """
    n = model["n"]
    vocab = sorted(model["vocab"])
    V = model["vocab_size"]
    counts = model["counts"]
    context_counts = model["context_counts"]
    prompt_toks = list(tokenize(prompt))

    def gen_context(seq):
        need = n - 1
        if len(seq) >= need:
            return tuple(seq[-need:])
        return tuple([BOS] * (need - len(seq)) + seq)

    def logp(ctx, w):
        p = (counts.get(ctx + (w,), 0) + 1) / (context_counts.get(ctx, 0) + V)
        return math.log(p)

    beams = [(0.0, list(prompt_toks))]
    completed = []
    for _ in range(max_length):
        candidates = []
        for score, seq in beams:
            ctx = gen_context(seq)
            for w in vocab:
                ns = score + logp(ctx, w)
                nseq = seq + [w]
                if w == EOS:
                    completed.append((ns, nseq[:-1]))
                else:
                    candidates.append((ns, nseq))
        if not candidates:
            break
        candidates.sort(key=lambda x: -x[0])
        beams = candidates[:beam_width]

    pool = completed + [(s, seq) for s, seq in beams]
    best_score, best_seq = max(pool, key=lambda x: x[0])
    return " ".join(best_seq)

````

````python
# --- Check your work: Q1.4 ---
model_bi = train_ngram(CORPUS, n=2)

# beam_width=1 must match greedy
greedy_out = greedy_decode(model_bi, "the cat", max_length=10)
beam1_out  = beam_decode(model_bi,   "the cat", beam_width=1, max_length=10)
print(f"greedy_decode: {greedy_out!r}")
print(f"beam_decode(width=1): {beam1_out!r}")
assert greedy_out == beam1_out, f"beam_width=1 must equal greedy: {greedy_out!r} != {beam1_out!r}"

# beam_width=3
beam3_out = beam_decode(model_bi, "the cat", beam_width=3, max_length=10)
print(f"beam_decode(width=3): {beam3_out!r}")
assert beam3_out.startswith("the cat"), "Output must start with prompt"

print("\nAll checks passed!")
````

### Q1.5 — `top_k_sample` (9 pts)

Implement `top_k_sample(model, prompt, k=5, max_length=20, rng=None) -> str`.

**What top-k sampling does:** it keeps some randomness, but only among the `k`
most likely next tokens under the current context.

**How to think about the implementation:**
- Score all tokens in the vocabulary.
- Sort or rank them by probability and keep only the top `k`.
- Re-normalise those `k` probabilities so they sum to 1.
- Sample one token from that smaller set using `rng.choice(...)`.
- Append it, update the context, and continue until `<EOS>` or `max_length`.

**Toy example:**
- Suppose the next-token probabilities are:
  `cat: 0.40`, `dog: 0.30`, `mat: 0.20`, `ran: 0.10`
- If `k=2`, only `cat` and `dog` remain.
- After re-normalisation, their probabilities become
  `cat: 0.40 / 0.70`, `dog: 0.30 / 0.70`.
- The model samples from those two options, so different random seeds can produce
  different outputs.

**Why re-normalise?** After restricting to top-k tokens, the remaining probabilities
no longer sum to 1, so we must divide by their total before sampling.

**Determinism:** `k=1` must always return the same token as `greedy_decode`
(argmax = only candidate after top-1 restriction).

**References:**
- Hugging Face text generation overview: https://huggingface.co/blog/how-to-generate#top-k-sampling
- Hugging Face generation strategies docs: https://huggingface.co/docs/transformers/main/en/generation_strategies

````python
def top_k_sample(model, prompt, k = 5, max_length = 20, rng = None):
    """
    Generate text from a prompt using top-k sampling.

    Args:
        model:      Trained model dict returned by train_ngram().
        prompt:     Input text string (will be tokenized).
        k:          Number of top-probability tokens to sample from.
        max_length: Maximum number of new tokens to generate.
        rng:        numpy RandomState for reproducibility; fresh state if None.

    Returns:
        Full string: original prompt tokens + generated tokens, space-joined.
    """
    if rng is None:
        rng = np.random.RandomState()
    n = model["n"]
    output = list(tokenize(prompt))
    vocab = sorted(model["vocab"])
    V = model["vocab_size"]
    counts = model["counts"]
    context_counts = model["context_counts"]

    def gen_context(seq):
        need = n - 1
        if len(seq) >= need:
            return tuple(seq[-need:])
        return tuple([BOS] * (need - len(seq)) + seq)

    for _ in range(max_length):
        ctx = gen_context(output)
        scored = [(w, (counts.get(ctx + (w,), 0) + 1) / (context_counts.get(ctx, 0) + V)) for w in vocab]
        scored.sort(key=lambda x: -x[1])
        top = scored[:k]
        total_p = sum(p for _, p in top)
        probs = np.array([p / total_p for _, p in top])
        words = [w for w, _ in top]
        choice = rng.choice(words, p=probs)
        if choice == EOS:
            break
        output.append(choice)
    return " ".join(output)

````

# Part 2: Attention Mechanism (20 points)

The **attention mechanism** (Vaswani et al., 2017 — *Attention Is All You Need*) is
the core computation inside every Transformer. Given an input sequence
$X \in \mathbb{R}^{T \times d_{\text{model}}}$, the **scaled dot-product attention** computes:

$$\text{Attention}(Q, K, V) = \text{softmax}\!\left(\frac{QK^\top}{\sqrt{d_k}}\right) V$$

where $Q = XW_Q$, $K = XW_K$, $V = XW_V$ are the Query, Key, and Value projections.

The $\frac{1}{\sqrt{d_k}}$ scaling prevents the dot products from becoming too large
as $d_k$ grows, which would push softmax into regions with vanishing gradients.

### Causal masking

For **auto-regressive** generation (like GPT), each position may only attend to
**earlier** positions. This is enforced by setting future attention scores to $-\infty$
before softmax:

$$\text{score}[i, j] = -\infty \quad \text{if } j > i$$

You will implement the three building blocks of single-head attention using **NumPy only**.

### Q2.1 — `compute_qkv` (6 pts)

Implement `compute_qkv(X, W_Q, W_K, W_V) -> (Q, K, V)`.

Given input embeddings $X \in \mathbb{R}^{T \times d_{\text{model}}}$ and three weight matrices,
compute the **Query**, **Key**, and **Value** projections:

$$Q = X W_Q, \quad K = X W_K, \quad V = X W_V$$



````python
def compute_qkv(X, W_Q, W_K, W_V):
    """
    Compute Query, Key, and Value matrices from input embeddings.

    Args:
        X:   Input embeddings of shape (seq_len, d_model).
        W_Q: Query weight matrix of shape (d_model, d_k).
        W_K: Key weight matrix of shape (d_model, d_k).
        W_V: Value weight matrix of shape (d_model, d_v).

    Returns:
        Q: Query matrix of shape (seq_len, d_k).
        K: Key matrix of shape (seq_len, d_k).
        V: Value matrix of shape (seq_len, d_v).
    """
    Q = X @ W_Q
    K = X @ W_K
    V = X @ W_V
    return Q, K, V
````

### Q2.2 — `compute_attention_scores` (6 pts)

Implement `compute_attention_scores(Q, K, causal=False) -> np.ndarray`.

Compute the **scaled** attention score matrix:

$$\text{scores} = \frac{Q K^\top}{\sqrt{d_k}}$$

If `causal=True`, apply a **causal mask**: set `scores[i, j] = -1e9` for all `j > i`
(future positions). This ensures that position $i$ cannot attend to position $j > i$.



````python
def compute_attention_scores(Q, K, causal = False):
    """
    Compute scaled attention scores QK^T / sqrt(d_k).

    Args:
        Q:      Query matrix of shape (seq_len, d_k).
        K:      Key matrix of shape (seq_len, d_k).
        causal: If True, mask future positions with -1e9.

    Returns:
        Attention score matrix of shape (seq_len, seq_len).
    """
    d_k = Q.shape[1]
    out = (Q @ K.T) / np.sqrt(d_k)
    if causal:
        out = np.copy(out)
        i, j = np.triu_indices_from(out, k=1)
        out[i, j] = -1e9
    return out
````

````python
# --- Check your work: Q2.2 ---
np.random.seed(345)
seq_len, d_k = 4, 4
Q_s = np.random.randn(seq_len, d_k)
K_s = np.random.randn(seq_len, d_k)
scores = compute_attention_scores(Q_s, K_s, causal=False)
print(f"Scores shape: {scores.shape}  (expected ({seq_len}, {seq_len}))")
assert scores.shape == (seq_len, seq_len)

# Scaling check: should match manual Q @ K.T / sqrt(d_k)
manual = Q_s @ K_s.T / np.sqrt(d_k)
assert np.allclose(scores, manual), "Scaling error"

# Causal mask: upper triangle should be very negative
scores_c = compute_attention_scores(Q_s, K_s, causal=True)
assert scores_c[0, 1] < -1e8, "Future positions should be masked to -1e9"
assert scores_c[1, 0] > -1e8, "Past positions should be unmasked"
print("Causal mask applied correctly ✓")
print("All checks passed!")
````

### Q2.3 — `scaled_dot_product_attention` (8 pts)

Implement `scaled_dot_product_attention(Q, K, V, causal=False) -> (output, weights)`.

Complete the full single-head attention computation:
1. Call `compute_attention_scores(Q, K, causal)` to get raw scores.
2. Apply **row-wise softmax** to convert scores to attention weights
   (use the numerically stable form: subtract row max before exponentiating).
3. Compute the final output.

**Returns:** `(output, attention_weights)` where:
- `output` has shape `(T, d_v)`
- `attention_weights` has shape `(T, T)` and each row sums to 1.


````python
def scaled_dot_product_attention(Q, K, V, causal = False):
    """
    Compute scaled dot-product attention.

    Args:
        Q:      Query matrix of shape (seq_len, d_k).
        K:      Key matrix of shape (seq_len, d_k).
        V:      Value matrix of shape (seq_len, d_v).
        causal: If True, apply causal mask.

    Returns:
        output:          Attention output of shape (seq_len, d_v).
        attention_weights: Row-wise softmax weights of shape (seq_len, seq_len).
    """
    scores = compute_attention_scores(Q, K, causal)
    m = np.max(scores, axis=1, keepdims=True)
    e = np.exp(scores - m)
    attention_weights = e / np.sum(e, axis=1, keepdims=True)
    output = attention_weights @ V
    return output, attention_weights
````

## K-Means Algorithm

K-Means partitions $n$ points into $K$ clusters by iterating until convergence:

1. **Initialise** centroids by randomly selecting $K$ data points.
2. **Assign** each point to the nearest centroid (Euclidean distance).
3. **Update** each centroid to the mean of its assigned points.
4. **Repeat** steps 2–3 until assignments do not change (or `max_iters` is reached).

Objective (Within-Cluster Sum of Squares / WCSS):

$$J = \sum_{k=1}^{K} \sum_{x_i \in C_k} \|x_i - \mu_k\|^2$$

### Q3.0 — `extract_embeddings` (6 pts)

Implement `extract_embeddings(texts, gpt2_tokenizer, gpt2_model, device, max_length=64) -> np.ndarray`.

- Run the model in `torch.no_grad()` mode.
- Mean-pool the **final hidden state** across the token dimension.
- Return a NumPy array of shape `(len(texts), hidden_size)`.

Hint: You may also refer to the code workbook for Lecture 18 to understand more about extracting embeddings using a model tokenizer in HuggingFace.


After implementing, run the check cell to construct:

````python
def extract_embeddings(texts, gpt2_tokenizer, model, device, max_length = 64):
    """Return mean-pooled DistilGPT-2 embeddings for a list of texts."""
    enc = gpt2_tokenizer(
        texts,
        padding=True,
        truncation=True,
        max_length=max_length,
        return_tensors="pt",
    )
    enc = {k: v.to(device) for k, v in enc.items()}
    with torch.no_grad():
        out = model(**enc)
    hidden = out.last_hidden_state
    mask = enc["attention_mask"].unsqueeze(-1).float()
    summed = (hidden * mask).sum(dim=1)
    lengths = mask.sum(dim=1).clamp(min=1e-9)
    pooled = (summed / lengths).cpu().numpy()
    return pooled
````

### Q3.2 — `assign_clusters` (6 pts)

Implement `assign_clusters(X, centroids) -> np.ndarray`.

For each data point $x_i$, assign it to the nearest centroid by **Euclidean distance**:

$$\text{assignment}[i] = \arg\min_k \| x_i - \mu_k \|_2^2$$

**Requirements:**
- Compute squared distances efficiently using NumPy broadcasting:
  `diffs = X[:, np.newaxis, :] - centroids[np.newaxis, :, :]` → shape `(n, K, d)`.
- Sum over the last axis and take `argmin` over `K`.
- Return a 1D integer array of shape `(n_samples,)`.


````python
def assign_clusters(X, centroids):
    """
    Assign each data point to the nearest centroid.

    Args:
        X:         Data matrix of shape (n_samples, n_features).
        centroids: Centroid matrix of shape (K, n_features).

    Returns:
        Cluster assignments: 1D array of shape (n_samples,), values in {0,...,K-1}.
    """
    diffs = X[:, np.newaxis, :] - centroids[np.newaxis, :, :]
    dists = np.sum(diffs ** 2, axis=2)
    return np.argmin(dists, axis=1)
````

### Q3.3 — `update_centroids` (6 pts)

Implement `update_centroids(X, assignments, K) -> np.ndarray`.

Update each centroid to the **mean** of all data points currently assigned to it:

$$\mu_k = \frac{1}{|C_k|} \sum_{x_i \in C_k} x_i$$

**Requirements:**
- For each cluster `k` in `range(K)`, compute the mean of `X[assignments == k]`.
- If a cluster is empty, keep its centroid as the zero vector (this is rare in practice).
- Return a `(K, n_features)` array.


````python
def update_centroids(X, assignments, K):
    """
    Update centroids as the mean of assigned data points.

    Args:
        X:           Data matrix of shape (n_samples, n_features).
        assignments: Cluster assignments of shape (n_samples,).
        K:           Number of clusters.

    Returns:
        New centroids of shape (K, n_features).
    """
    d = X.shape[1]
    centroids = np.zeros((K, d))
    for k in range(K):
        m = assignments == k
        if np.any(m):
            centroids[k] = X[m].mean(axis=0)
    return centroids
````

### Q3.4 — `kmeans` (8 pts)

Implement `kmeans(X, K, max_iters=100, seed=None) -> (centroids, assignments, inertia)`.

Put together the full K-Means loop:
1. Initialise centroids with `initialize_centroids`.
2. Repeat up to `max_iters` times:
   a. Assign each point to the nearest centroid (`assign_clusters`).
   b. If assignments did not change from the previous iteration, **break** (convergence).
   c. Update centroids (`update_centroids`).
3. Compute **inertia** (WCSS):
   $J = \sum_{k=0}^{K-1} \sum_{x_i \in C_k} \|x_i - \mu_k\|^2$
4. Return `(centroids, assignments, inertia)`.


````python
def kmeans(X, K, max_iters = 100, seed = None):
    """
    Run the K-Means algorithm.

    Args:
        X:         Data matrix of shape (n_samples, n_features).
        K:         Number of clusters.
        max_iters: Maximum number of iterations.
        seed:      Random seed for centroid initialisation.

    Returns:
        centroids:   Final centroids of shape (K, n_features).
        assignments: Final cluster assignments of shape (n_samples,).
        inertia:     Within-cluster sum of squares (float).
    """
    centroids = initialize_centroids(X, K, seed)
    prev = None
    for _ in range(max_iters):
        assignments = assign_clusters(X, centroids)
        if prev is not None and np.array_equal(assignments, prev):
            break
        centroids = update_centroids(X, assignments, K)
        prev = assignments.copy()
    inertia = float(np.sum((X - centroids[assignments]) ** 2))
    return centroids, assignments, inertia
````

### Q3.5 — `compute_elbow_data` (6 pts)

Implement `compute_elbow_data(X, K_max, seed=42) -> List[float]`.

Run K-Means for every `K` from 1 to `K_max` (inclusive) and collect the inertia.
The resulting list is used to plot the **elbow curve** — the "elbow" visually suggests
the best number of clusters.

**Requirements:**
- Return a list of length `K_max`: `[inertia_1, inertia_2, ..., inertia_K_max]`.
- Inertia must be **monotonically non-increasing** (adding clusters can only reduce WCSS).


````python
def compute_elbow_data(X, K_max, seed = 42):
    """
    Compute inertia for K = 1 to K_max.

    Args:
        X:     Data matrix.
        K_max: Maximum number of clusters to evaluate.
        seed:  Random seed passed to kmeans().

    Returns:
        List of inertia values [inertia_K1, ..., inertia_Kmax].
    """
    return [kmeans(X, k, seed=seed)[2] for k in range(1, K_max + 1)]
````

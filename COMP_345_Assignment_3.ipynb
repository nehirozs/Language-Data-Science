### 1.1 `linear_transform` (2 points)

Use NumPy (already loaded as `np`) to calculate the linear transform.

This function computes a single linear transformation:

$$
z = \sum_i x_i w_i + b
$$

Arguments:
- `x (List[float])`: Input features  
- `w (List[float])`: Weights (must be the same length as `x`)  
- `b (float)`: Bias term  

Returns:
- `float`: The computed scalar value \( z \)

Examples:
```python
>>> x = [2.0, 3.0]
>>> w = [4.0, 5.0]
>>> b = 1.0
>>> linear_transform(x, w, b)
24.0
```

````python

def linear_transform(x: List[float], w: List[float], b: float) -> float:
    ### WRITE YOUR CODE BELOW
    z = sum(xi * wi for xi, wi in zip(x, w)) + b
    ### END CODE HERE
    return z
````

### 1.3 `perceptron_training_step` (4 points)

Implement **one update step** of the perceptron learning algorithm for a single training example.

Given:
- `x_i`: input feature vector  
- `y_true`: true label (0 or 1)  
- `w`: current weight vector  
- `b`: current bias  
- `lr`: learning rate  

#### Steps

1. Compute the predicted label using your previously implemented function:  
   `y_pred = perceptron_predict(x_i, w, b)`

2. Compute the prediction error:  
   `error = y_true - y_pred`

3. If the prediction is incorrect (`error != 0`):
   - Update each weight using  
     `w_i = w_i + lr * error * x_i`
   - Update the bias using  
     `b = b + lr * error`

4. If the prediction is correct (`error == 0`), leave the weights and bias unchanged.

#### Return

Return a tuple `(new_w, new_b, error)` where:
- `new_w` is the updated list of weights  
- `new_b` is the updated bias  
- `error` is the computed prediction error

#### Example

```python
>>> x_i = [2.0, 1.0]
>>> y_true = 1
>>> w = [0.0, 0.0]
>>> b = -1.0
>>> lr = 0.1
>>> perceptron_training_step(x_i, y_true, w, b, lr)
([0.2, 0.1], -0.9, 1)
```

````python

def perceptron_training_step(x_i: List[float], y_true: int, w: List[float], b: float, lr: float) -> Tuple[List[float], float, int]:
    ### WRITE YOUR CODE BELOW
    y_pred = perceptron_predict(x_i, w, b)
    error = y_true - y_pred
    new_w = list(w)
    new_b = b
    if error != 0:
        for i in range(len(new_w)):
            new_w[i] = w[i] + lr * error * x_i[i]
        new_b = b + lr * error
    ### END CODE HERE
    return new_w, new_b, error
````

### 1.4 `relu_vector` (2 points)

Implement the ReLU (Rectified Linear Unit) function **element-wise** to an input vector.

The ReLU function is defined as:

$$
\mathrm{ReLU}(z) = \max(0, z)
$$

This means:
- If a value is positive, keep it unchanged.
- If a value is negative, replace it with 0.
- If a value is 0, it remains 0.

Your function should take a list of numbers and return a new list where ReLU has been applied to each element individually.

#### Example

```python
>>> relu_vector([-1, 0, 2])
>>> [0.0, 0.0, 2.0]
```

````python

def relu_vector(z: List[float]) -> List[float]:
    ### WRITE YOUR CODE BELOW
    h = [max(0.0, zi) for zi in z]
    ### END CODE HERE
    return h
````

### 1.5 `mlp_forward_one_hidden` (4 points)

Implement **one forward pass** for a neural network with **one hidden layer**.

The network has:
- Input vector `x`
- First layer weights `W1` and bias `b1`
- Second layer weights `W2` and bias `b2`
- ReLU activation applied after the first layer

The forward computation follows these steps:

1. Compute the first linear transformation:
   $$
   z_1 = W_1 x + b_1
   $$

2. Apply the ReLU activation element-wise:
   $$
   h_1 = \mathrm{ReLU}(z_1)

3. Compute the second linear transformation:
   $$
   z_2 = W_2 h_1 + b_2
   $$

Do **not** update any weights or biases. This function performs computation only.

#### Return

Return a dictionary with the following keys:

- `"z1"`: the pre-activation vector of the hidden layer  
- `"h1"`: the hidden layer activations after ReLU

The returned structure should look like:

#### Example


````python

def mlp_forward_one_hidden(x: List[float], W1: List[List[float]], b1: List[float], W2: List[List[float]], b2: List[float]) -> Dict[str, List[float]]:
    ### WRITE YOUR CODE BELOW
    z1 = [sum(W1[j][i] * x[i] for i in range(len(x))) + b1[j] for j in range(len(W1))]
    h1 = relu_vector(z1)
    z2 = [sum(W2[0][j] * h1[j] for j in range(len(h1))) + b2[0]]
    states = {"z1": z1, "h1": h1, "z2": z2}
    ### END CODE HERE
    return states
````

### 1.6 `softmax` (3 points)

Implement a **softmax** function that converts a list of logits
(real-valued scores) into probabilities.

The softmax function is defined as:

$$
\mathrm{softmax}(z_i) = \frac{e^{z_i}}{\sum_j e^{z_j}}
$$

This ensures:
- All output values are non-negative.
- The outputs sum to 1.
- The result can be interpreted as a probability distribution.

#### Example

```python
>>> softmax([2.0, 1.0, 0.1])
>>> [0.659..., 0.242..., 0.099...]
```

The probabilities sum to 1.

Your function should return a list of probabilities of the same length as the input.

````python

def softmax(logits: List[float]) -> List[float]:
    ### WRITE YOUR CODE BELOW
    exp_z = [math.exp(zi) for zi in logits]
    total = sum(exp_z)
    probs = [e / total for e in exp_z]
    ### END CODE HERE
    return probs
````

### 1.7 `cross_entropy_loss` (3 points)

Implement the cross-entropy loss for a single example using the predicted
probabilities and the index of the true class.

For a given probability distribution `probs` and true class index `true_class`,
the loss is defined as:

$$
\mathcal{L} = -\log\big(p_{\text{true class}}\big)
$$

#### Steps

1. Select the probability for the true class:  
   `p = probs[true_class]`

2. Set `epsilon = 1e-12` for numerical stability.

3. Clip the probability to avoid log(0):  
   `p_clipped = max(p, epsilon)`

4. Compute the loss as `−log(p_clipped)`.

5. Return the loss as a Python float.

#### Example

```python
>>> cross_entropy_loss([0.1, 0.7, 0.2], 1)
>>> 0.35667494393873245
```

Your function should return a single float representing the loss value.

````python
def cross_entropy_loss(probs: List[float], true_class: int) -> float:
    ### WRITE YOUR CODE BELOW
    p = probs[true_class]
    epsilon = 1e-12
    p_clipped = max(p, epsilon)
    loss = -math.log(p_clipped)
    ### END CODE HERE
    return loss
````

### 2.3 `compute_tf` (4 points)

Implement **Term frequency (TF)** normalization.

The term frequency of a word in a document is the count of that word divided by the document length:

$$
TF(t, d) = \frac{\text{count}}{\text{doc\_length}}
$$

Arguments:
- `count` (int): number of times the word appears in the document
- `doc_length` (int): total number of tokens in the document

Requirements:
- If `doc_length <= 0`, return `0.0`.
- If `count <= 0`, return `0.0`.
- Return a Python `float`.

#### Example

````python
def compute_tf(count: int, doc_length: int) -> float:
    ### WRITE YOUR CODE BELOW
    if doc_length <= 0 or count <= 0:
        tf = 0.0
    else:
        tf = count / doc_length
    ### END CODE HERE
    return tf
````

### 2.4 `compute_idf_smooth` (4 points)

Compute the **smoothed inverse document frequency (IDF)** for a given word.

The function takes `count_df` (a word–document count DataFrame): rows = words, columns = documents, values = counts.

Use the following formula:

$$
IDF = \log\left(\frac{N + 1}{df + 1}\right) + 1
$$

Where:
- `N` is the total number of documents (number of columns in `count_df`)
- `df` = document frequency = number of documents in which the word appears at least once (count > 0).

#### Steps

1. Set `N = count_df.shape[1]` (number of documents).
2. Get the row for `word` in `count_df`; count how many columns have value > 0 → that is `df`.
3. Compute `idf = math.log((N + 1) / (df + 1)) + 1`.
4. Return `idf` as a float.

#### Example of `count_df`

|        | DocA | DocB |
|--------|------|------|
| apple  |  0   |  2   |
| banana |  1   |  1   |
| carrot |  1   |  0   |

#### Example

```python
>>> count_df = pd.DataFrame({"DocA": [0, 1, 1], "DocB": [2, 1, 0]}, index=["apple", "banana", "carrot"])
>>> compute_idf_smooth("apple", count_df)
1.4054651081081644
>>> compute_idf_smooth("banana", count_df)
1.0
```

````python
def compute_idf_smooth(word: str, count_df: pd.DataFrame) -> float:
    ### WRITE YOUR CODE BELOW
    N = count_df.shape[1]
    row = count_df.loc[word]
    df = (row > 0).sum()
    idf = math.log((N + 1) / (df + 1)) + 1
    ### END CODE HERE
    return idf
````

### 2.5 `compute_tfidf_matrix` (4 points)

Convert `count_df` (a word–document count DataFrame) into a TF-IDF DataFrame.

The function takes `count_df` (a word–document count DataFrame, see example).

Formula: TF-IDF = TF × IDF

#### Steps

1. You receive `count_df` as input (rows = words, columns = documents, values = counts).

2. Compute **TF** using `compute_tf(count, doc_length)` (Section 2.3):
   - `count` = the value in cell `count_df.at[word, doc]` (how many times the word appears in that document)
   - `doc_length` = sum of all counts in that document column: `count_df[doc].sum()`

3. Compute **IDF** using `compute_idf_smooth(word, count_df)` (Section 2.4):
   - Inputs: `word` (the row label) and `count_df` (the same DataFrame you received)
   - Returns a single float per word

4. For each cell (word, document): multiply TF × IDF and place in the result DataFrame. Return a DataFrame with the same rows and columns as `count_df`.

Hint: This function should finish in at most a couple of minutes. Avoid deeply nested Python `for` loops over the full DataFrame; use NumPy/Pandas vectorized operations where possible.

#### Example

```python
>>> count_df = pd.DataFrame({"DocA": [0, 1, 1], "DocB": [2, 1, 0]}, index=["apple", "banana", "carrot"])
>>> compute_tfidf_matrix(count_df)
            DocA      DocB
apple   0.000000  0.936977
banana  0.500000  0.333333
carrot  0.702733  0.000000
```

````python
def compute_tfidf_matrix(count_df: pd.DataFrame) -> pd.DataFrame:
    ### WRITE YOUR CODE BELOW
    doc_lengths = count_df.sum(axis=0)
    tfidf_df = pd.DataFrame(index=count_df.index, columns=count_df.columns, dtype=float)
    for word in count_df.index:
        idf = compute_idf_smooth(word, count_df)
        for doc in count_df.columns:
            count = count_df.at[word, doc]
            doc_len = doc_lengths[doc]
            tf = compute_tf(int(count), int(doc_len))
            tfidf_df.at[word, doc] = tf * idf
    ### END CODE HERE
    return tfidf_df
````

### 3.1 `cosine_similarity_manual` (4 points)

Implement **cosine similarity** for two vectors using NumPy.

Use `np.dot()` for the dot product and `np.linalg.norm()` for the L2 norm. Do not implement the dot product or norm manually.

Cosine similarity between two vectors:

$$
\text{cosine-similarity}(a, b) =
\frac{a \cdot b}{\|a\|_2 \, \|b\|_2}
$$

#### Steps

1. Convert inputs to NumPy arrays (e.g., `np.array(vec_a)`).
2. Compute dot product using: `np.dot()`.
3. Compute L2 norms using: `np.linalg.norm()`.
4. If either norm is 0, return `0.0`.
5. Otherwise return dot_product / (norm_a * norm_b) as a float.

#### Example

```python
>>> cosine_similarity_manual([1, 0, 1], [1, 1, 0])
0.5
>>> cosine_similarity_manual([1, 0], [1, 0])
1.0
>>> cosine_similarity_manual([0, 0], [1, 0])
0.0
```

````python
def cosine_similarity_manual(vec_a: List[float], vec_b: List[float]) -> float:
    ### WRITE YOUR CODE BELOW
    a = np.array(vec_a)
    b = np.array(vec_b)
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    if norm_a == 0 or norm_b == 0:
        sim = 0.0
    else:
        sim = float(dot_product / (norm_a * norm_b))
    ### END CODE HERE
    return sim
````

````python
def build_cooccurrence_matrix(tokens: List[str], vocab: List[str], window_size: int) -> pd.DataFrame:
    ### WRITE YOUR CODE BELOW
    vocab_set = set(vocab)
    cooccur_df = pd.DataFrame(0, index=vocab, columns=vocab, dtype=int)
    n = len(tokens)
    for i in range(n):
        center = tokens[i]
        if center not in vocab_set:
            continue
        for j in range(max(0, i - window_size), min(n, i + window_size + 1)):
            if j == i:
                continue
            context = tokens[j]
            if context in vocab_set:
                cooccur_df.at[center, context] += 1
    ### END CODE HERE
    return cooccur_df
````

### 3.4 `rank_similar_words` (4 points)

Rank neighboring words for a given target word using a **word–word similarity matrix**.

**Input:**
- `target`: the word to find neighbors for.
- `sim_df`: square DataFrame (rows = words, columns = words, values = cosine similarity scores), e.g. from Section 3.3.
- `k`: maximum number of neighbors to return.
- `most_similar`: if True, return highest-similarity words first; if False, return lowest first.

**Requirements:**
- Return `[]` if target is not in `sim_df` or if `k <= 0`.
- Get the target row; exclude the target word itself.
- Sort by similarity: highest first if `most_similar=True`, lowest first if `False`. For ties, sort by word A–Z.
- Return the first `k` as a list of `(word, similarity)` tuples.

**Example 1 (most similar):**

```python
>>> sim_df = get_cooccurrence_cosine_similarity_matrix(cooccur_df)  # from Section 3.3
>>> rank_similar_words("bag", sim_df, k=2, most_similar=True)
[("dog", 0.777), ("car", 0.449)]
```

**Example 2 (most dissimilar):**

```python
>>> rank_similar_words("bag", sim_df, k=2, most_similar=False)
[("car", 0.449), ("dog", 0.777)]
```

**Note:** `sim_df` is provided as input; you do not need to compute it. The examples above assume `sim_df` is already available (e.g. from Section 3.3).

````python
def rank_similar_words(target: str, sim_df: pd.DataFrame, k: int, most_similar: bool = True) -> List[Tuple[str, float]]:
    ### WRITE YOUR CODE BELOW
    if target not in sim_df.index or k <= 0:
        return []
    row = sim_df.loc[target].drop(target, errors="ignore")
    pairs = [(word, float(sim)) for word, sim in row.items()]
    pairs.sort(key=lambda p: (-p[1], p[0]) if most_similar else (p[1], p[0]))
    ranked = pairs[:k]
    ### END CODE HERE
    return ranked
````

### 3.5 `top_context_neighbors` (4 points)

For a given `target` word, return the top-`k` context neighbors from `cooccur_df` as a list of `(word, count)` tuples.


This is a basic way to inspect distributional meaning: words with similar neighbors are often semantically related.
In real NLP pipelines, this helps with quick vocabulary analysis, query expansion, and sanity-checking learned representations.

**For example:**
In a customer-support chatbot corpus, if the target word is `refund`, strong neighbors like `payment`, `order`, and `cancel` indicate that refund-related intent and language are clustered correctly.

**Inputs:**
- `cooccur_df`: a word-word co-occurrence matrix (`pandas.DataFrame`)
  - index = target words
  - columns = context words
  - values = integer co-occurrence counts within a context window
- `target`: the row word whose context neighborhood you want
- `k`: maximum number of neighbors to return

**Requirements:**
- Use the row corresponding to `target`.
- If `target` is not in the DataFrame index, return an empty list.
- Consider only entries with count > 0.
- Sort by:
  1) Count in descending order
  2) Word in alphabetical order (for ties)
- Return at most `k` neighbors.
- Counts must be integers.

**Example 1** (coffee, morning, breakfast — words that often appear together; football — unrelated):

```python
>>> cooccur_df
```


````python

def top_context_neighbors(target: str, cooccur_df: pd.DataFrame, k: int) -> List[Tuple[str, int]]:
    ### WRITE YOUR CODE BELOW
    if target not in cooccur_df.index:
        return []
    row = cooccur_df.loc[target]
    pairs = [(word, int(count)) for word, count in row.items() if count > 0]
    pairs.sort(key=lambda p: (-p[1], p[0]))
    neighbors = pairs[:k]
    ### END CODE HERE
    return neighbors
````

### 4.2 `solve_analogy` (10 points)

Use the pre-trained Word2Vec model to solve the analogy `a : b :: c : ?`.

**Inputs:**
- `a`, `b`, `c`: words for the analogy `a : b :: c : ?`
    -- `word2vec_model`: the pre-trained model (loaded as in Section 4.1 with `api.load('word2vec-google-news-300')`)
- `top_k`: number of answers to return

**Output:**
- A list of `(word, similarity)` tuples (floats), sorted from most similar to least similar.

**Hint:**
- You can use the model’s built-in nearest-neighbor search instead of looping over the entire vocabulary:
- `word2vec_model.most_similar(positive=[b, c], negative=[a], topn=...)`
- If `top_k <= 0`, return `[]`.
- If any of `a`, `b`, or `c` is out of vocabulary (not in `word2vec_model.key_to_index`), return `[]`.
- Exclude `a`, `b`, and `c` from candidates. To still return up to `top_k` answers, over-fetch (e.g., request more than `top_k`), filter out excluded words, then take the first `top_k` remaining items.

**Example (Python):**

```python
>>> solve_analogy('man', 'king', 'woman', word2vec_model, top_k=1)
[('queen', 0.73)]  # similarity score shown here is just an example

>>> solve_analogy('france', 'paris', 'italy', word2vec_model, top_k=1)
[('rome', 0.80)]   # similarity score shown here is just an example
```

**Simple hint:**
- First check words with `word2vec_model.key_to_index`.
- Then use `word2vec_model.most_similar(...)` to find answers.


````python

def solve_analogy(a: str, b: str, c: str, word2vec_model, top_k: int = 3) -> List[Tuple[str, float]]:
    ### WRITE YOUR CODE BELOW
    if top_k <= 0:
        return []
    key_to_index = word2vec_model.key_to_index
    if a not in key_to_index or b not in key_to_index or c not in key_to_index:
        return []
    exclude = {a, b, c}
    raw = word2vec_model.most_similar(positive=[b, c], negative=[a], topn=top_k + len(exclude))
    ranked = [(w, s) for w, s in raw if w not in exclude][:top_k]
    ### END CODE HERE
    return ranked
````

````python
### DO NOT MODIFY THIS CELL ###

def clean_sentiment_dataframe(df: pd.DataFrame) -> pd.DataFrame:
    """
    Input:
        DataFrame with columns ['text', 'label'] where label is 0 or 1.
    Output:
        Cleaned DataFrame with same columns.
        - Only valid labels (0/1)
        - Cleaned text
        - No empty rows
    """

    # Validate required columns
    if "text" not in df.columns or "label" not in df.columns:
        raise ValueError("DataFrame must contain columns ['text', 'label'].")

    out = df.copy()

    # Ensure numeric labels and keep only 0/1
    out["label"] = pd.to_numeric(out["label"], errors="coerce")
    out = out[out["label"].isin([0, 1])].copy()
    out["label"] = out["label"].astype(int)

    #NLTK stopwords
    stop_words = set(stopwords.words("english"))

    # Keep negations (important for sentiment)
    negations = {"no", "nor", "not", "never", "n't"}
    stop_words -= negations

    # Precompiled regex
    url_re = re.compile(r"https?://\S+|www\.\S+", flags=re.IGNORECASE)
    email_re = re.compile(r"\b[\w\.-]+@[\w\.-]+\.\w+\b", flags=re.IGNORECASE)
    num_re = re.compile(r"\b\d+(\.\d+)?\b")
    multi_space_re = re.compile(r"\s+")

    def clean_text(x) -> str:
        if pd.isna(x):
            return ""

        s = str(x)
        # Normalize unicode
        s = unicodedata.normalize("NFKC", s)
        # Lowercase
        s = s.lower()
        # Replace URLs and emails
        s = url_re.sub(" <url> ", s)
        s = email_re.sub(" <email> ", s)
        # Remove numbers
        s = num_re.sub(" ", s)
        # Remove punctuation (keep apostrophes and placeholders)
        s = re.sub(r"[^\w\s'<>]+", " ", s)

        # Tokenize (deterministic)
        tokens = wordpunct_tokenize(s)

        cleaned_tokens = []
        for token in tokens:
            if token in {"<url>", "<email>"}:
                cleaned_tokens.append(token)
                continue
            if len(token) < 2:
                continue
            if token in stop_words:
                continue
            cleaned_tokens.append(token)

        s_out = " ".join(cleaned_tokens)
        s_out = multi_space_re.sub(" ", s_out).strip()

        return s_out

    # Apply cleaning
    out["text"] = out["text"].map(clean_text)

    # Remove empty rows
    out = out[out["text"].str.len() > 0].copy()
    out.reset_index(drop=True, inplace=True)
    return out


# Example:
# clean_df = clean_sentiment_dataframe(df)
````

````python
def train_sentiment_model() -> Dict[str, object]:
    ### WRITE YOUR CODE BELOW
    random.seed(SENTIMENT_SEED)
    np.random.seed(SENTIMENT_SEED)
    data_df = pd.read_csv(SENTIMENT_DATASET_CSV_URL)
    clean_df = clean_sentiment_dataframe(data_df)
    X = clean_df["text"]
    y = clean_df["label"]
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=SENTIMENT_TEST_SIZE, random_state=SENTIMENT_SEED, shuffle=True, stratify=y
    )
    vectorizer = TfidfVectorizer(
        lowercase=False,
        token_pattern=r"(?u)\b\w+\b",
        ngram_range=SENTIMENT_NGRAM_RANGE,
        min_df=SENTIMENT_MIN_DF,
        max_features=SENTIMENT_MAX_FEATURES,
        norm="l2",
    )
    X_train_vec = vectorizer.fit_transform(X_train)
    X_test_vec = vectorizer.transform(X_test)
    clf = MultinomialNB(alpha=1.0)
    clf.fit(X_train_vec, y_train)
    y_pred = clf.predict(X_test_vec)
    acc = round(accuracy_score(y_test, y_pred), 6)
    f1 = round(f1_score(y_test, y_pred, average="macro"), 6)
    cm = confusion_matrix(y_test, y_pred)
    report = {
        "accuracy": acc,
        "f1_macro": f1,
        "confusion_matrix": [[int(cm[0, 0]), int(cm[0, 1])], [int(cm[1, 0]), int(cm[1, 1])]],
    }
    ### END CODE HERE
    return report
````
